#### Select all the customers from Spain.
**Hint**: Remember that you are not asked to find orders from Spain but customers. A customer might have more than one order associated. 

```python
spanish_orders = orders[(orders["Country"] == "Spain")]

spanish_orders["CustomerID"]
```

#### How many customers do we have in Spain?

```python
spanish_orders["CustomerID"].nunique()
```

#### Select all the customers who have bought more than 50 items.
**Hint**: Remember that you are not asked to find orders with more than 50 items but customers who bought more than 50 items. A customer with two orders of 30 items each should appear in the selection.

```python
cust_50 = orders[orders["Quantity"] > 50]
cust_50 = cust_50["InvoiceNo"]

cust_50
```

#### Select orders from Spain that include more than 50 items.

```python
cust_50_spain = orders[(orders["Quantity"] > 50) & (orders["Country"] == "Spain")]
cust_50_spain = cust_50_spain["InvoiceNo"]

cust_50_spain
```

#### Select all free orders.

```python
free_orders = orders[orders["amount_spent"] == 0]

free_orders
```

#### Select all orders whose description starts with `lunch bag`.
**Hint**: use string functions.

```python
lunch_orders = orders[orders["Description"].str.startswith("lunch bag")]

lunch_orders
```

#### Select all `lunch bag` orders made in 2011.

```python
lunch_orders_2011 = orders[(orders["Description"].str.startswith("lunch bag")) & (orders["year"] == 2011)]

lunch_orders_2011
```

#### Show the frequency distribution of the amount spent in Spain.

```python
spanish_orders_freq_dist = spanish_orders["amount_spent"].value_counts()

spanish_orders_freq_dist
```

#### Select all orders made in the month of August.

```python
orders_august = orders[orders["month"] == 8]

orders_august
```

#### Find the number of orders made by each country in the month of August.
**Hint**: Use value_counts().

```python
country_orders_august = orders_august["Country"].value_counts()

country_orders_august
```

#### What's the  average amount of money spent by country?

```python
avg_spent_country = orders.groupby("Country")["amount_spent"].mean()

avg_spent_country
```

#### What's the most expensive item?

```python
expensive_prod = orders[orders["amount_spent"] == orders["amount_spent"].max()]

print(expensive_prod["Description"])
```

#### What is the average amount spent per year?

```python
avg_per_year = orders.groupby("year")["amount_spent"].mean()

avg_per_year
```
